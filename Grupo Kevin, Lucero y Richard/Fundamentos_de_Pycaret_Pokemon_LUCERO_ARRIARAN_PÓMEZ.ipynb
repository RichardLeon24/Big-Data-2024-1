# Paso 1: Cargamos Pycaret y un Dataset

```python
import pycaret
pycaret.__version__
```

```python
from pycaret.datasets import get_data
Datos_traffic = get_data(dataset = 'pokemon') #Aquí cambias el Dataset que te haya tocado.
```

# Paso 2: Importar las Regresiones de Pycaret y Crear el Objeto

```python
from pycaret.regression import *
```

```python
# import RegressionExperiment and init the class
from pycaret.regression import RegressionExperiment, finalize_model, save_model
ExperimentoRegresion = RegressionExperiment()
```

# Paso 3: Configurar el Experimento u Objeto

```python
# init setup on exp
ExperimentoRegresion.setup(Datos_traffic, target = 'Sp. Atk', session_id = 123)
```

# Paso 4: Entrenamos y Comparamos los Modelos

```python
#compare models using OOP
MejorModelo = ExperimentoRegresion.compare_models()
```

Además, su R2 es perfecto (1.000), lo que significa que el modelo explica completamente la variabilidad de la variable dependiente. Comparado con otros modelos, como el Huber Regressor, que también muestra un rendimiento notable, el Passive Aggressive Regressor aún sobresale por sus errores mínimos y su capacidad de predicción perfecta, lo que lo convierte en la opción más eficiente y precisa para las tareas de regresión evaluadas en esta tabla.
