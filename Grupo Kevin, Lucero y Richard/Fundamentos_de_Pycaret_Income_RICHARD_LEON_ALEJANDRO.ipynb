# Paso 1: Cargamos Pycaret y un Dataset

```python
import pycaret
pycaret.__version__
```

```python
from pycaret.datasets import get_data
Datos_traffic = get_data(dataset = 'income') #Aquí cambias el Dataset que te haya tocado.
```

# Paso 2: Importar las Regresiones de Pycaret y Crear el Objeto

```python
from pycaret.regression import *
```

```python
# import RegressionExperiment and init the class
from pycaret.regression import RegressionExperiment, finalize_model, save_model
ExperimentoRegresion = RegressionExperiment()
```

# Paso 3: Configurar el Experimento u Objeto

```python
# init setup on exp
ExperimentoRegresion.setup(Datos_traffic, target = 'capital-gain', session_id = 2174)
```

# Paso 4: Entrenamos y Comparamos los Modelos

```python
#compare models using OOP
MejorModelo = ExperimentoRegresion.compare_models()
```

### *Resolver*
#### 1. ¿Cuál es el mejor modelo y por qué? Justificar. ¿Qué significa cada métrica?

El Gradient Boosting Regressor (gbr) parece ser el mejor modelo basado en las siguientes observaciones:

*   Tiene el valor más bajo de MAE (1753.5620), lo que indica que en promedio sus predicciones son las más cercanas a los valores reales.
*   Presenta un valor de R2 positivo (0.0885), aunque no es muy alto, sugiere que el modelo explica una pequeña parte de la variabilidad de la variable dependiente.
*   El RMSE (7186.8354) es relativamente bajo en comparación con otros modelos.
*   El tiempo de entrenamiento (0.7118 segundos) es razonablemente corto.

Aunque el Huber Regressor tiene un valor de RMSLE (2.5373) menor, su MAE, MSE y RMSE son más altos, y su R2 es negativo (-0.0220), lo que sugiere un peor ajuste en general. Por lo tanto, el Gradient Boosting Regressor se considera el mejor modelo en este caso.

En conclusion, el Gradient Boosting Regressor se destaca por su baja MAE, su R2 relativamente alto, un buen RMSE y un tiempo de entrenamiento eficiente. Este conjunto equilibrado de características lo hace el mejor modelo en términos de precisión, capacidad explicativa y eficiencia computacional.

### *Resolver*
#### 2. ¿Por qué escogiste estos 7 modelos en la lista de include?. Si deseas puedes ayudarte del ordenamiento.

Escogi estos 7 métodos porque cada uno tiene ventajas específicas que pueden ser útiles en diferentes escenarios de análisis de datos y tambien por el orden de lista que a mi parecer destaca por ser primeros:

* **LR:** Por su simplicidad y facilidad de interpretación cuando la relación entre variables es lineal.

* **Lasso**: Para realizar selección de variables y simplificar el modelo en presencia de muchas características.

* **Ridge**: Para manejar la multicolinealidad y prevenir el sobreajuste sin descartar variables.

* **en:** Para combinar las ventajas de Lasso y Ridge cuando hay muchas variables correlacionadas.

* **LAR**: Para una selección de variables rápida y eficiente en problemas de alta dimensionalidad.

* **LLAR:** Para aprovechar la selección de variables de Lasso con la velocidad de LAR.

* **OMP**: Para identificar iterativamente las variables más importantes en análisis esparso y problemas de compresión de señales.
