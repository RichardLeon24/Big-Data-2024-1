# Paso 1: Cargamos Pycaret y un Dataset

```python
import pycaret
pycaret.__version__
```

```python
from pycaret.datasets import get_data
Datos_traffic = get_data(dataset = 'population') #Aquí cambias el Dataset que te haya tocado.
```

# Paso 2: Importar las Regresiones de Pycaret y Crear el Objeto

```python
from pycaret.regression import *
```

```python
# import RegressionExperiment and init the class
from pycaret.regression import RegressionExperiment, finalize_model, save_model
ExperimentoRegresion = RegressionExperiment()
```

# Paso 3: Configurar el Experimento u Objeto

```python
# init setup on exp
ExperimentoRegresion.setup(Datos_traffic, target = '1961', session_id = 2174)
```

# Paso 4: Entrenamos y Comparamos los Modelos

```python
#compare models using OOP
MejorModelo = ExperimentoRegresion.compare_models()
```

### *Resolver*
#### 1. ¿Cuál es el mejor modelo y por qué? Justificar. ¿Qué significa cada métrica?

El Orthogonal Matching Pursuit (omp) parece ser el mejor modelo basado en las siguientes observaciones:

*   Tiene el valor más bajo de MAE (0.2247), lo que indica que en promedio sus predicciones son las más cercanas a los valores reales.
*   Presenta un valor de R2 positivo (0.8527), aunque es el mas alto, sugiere que el modelo explica una pequeña parte de la variabilidad de la variable dependiente.
*   El RMSE (0.4117) es relativamente bajo en comparación con otros modelos.

En conclusion, el Orthogonal Matching Pursuit	 se destaca por su R2 relativamente alto, un buen RMSE y un tiempo de entrenamiento eficiente. Este conjunto equilibrado de características lo hace el mejor modelo en términos de precisión, capacidad explicativa y eficiencia computacional.
